```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)

data = np.random.rand(100)
df = pd.DataFrame({"random_numbers": data})

df.head()
```

```python
mean_val = df["random_numbers"].mean()
std_val = df["random_numbers"].std()

summary_df = pd.DataFrame({
    "mean": [mean_val],
    "std": [std_val]
})

summary_df.to_csv(r"C:\Users\Usuario\Desktop\tp0-test-env-aguselara1-tech\results/test_output.csv", index=False)

summary_df
```
